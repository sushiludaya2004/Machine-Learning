```python
import pandas as pd
import numpy as np
```

```python
ts = pd.Series(np.random.randn(1000),
               index=pd.date_range('1/1/2000', periods=1000))
```

```python
ts = ts.cumsum()
```

```python
ts.head()
```

#### Line chart

```python
ts.plot()
```
